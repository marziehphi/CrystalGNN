# Import Libraries

```python
import numpy as np
import pandas as pd

from IPython import display
import matplotlib.pyplot as plt

import os
import glob
import json
import csv
```

# Extracting CIFs file

```python
id = 'X'
id_a = 'Y' 
id_b = 'Z'

opt_cifs = [f for f in glob.glob(f'database/{id_a}/{id_b}/CSP/{id}/*.cif')] # Accurate CIFS
all_cifs = [f for f in glob.glob(f'database/{id_a}/{id_b}/CSP/sorted_structures/*.cif')] # Intermediate CIFs

idx = []
all_idx = []

for file in range(len(opt_cifs)):
    name, ext = os.path.splitext(opt_cifs[file])
    name = name.split('/')
    #name = name[5].split('_')
    name = name[5].split('_')
    idx.append(name[1])
    
for file in range(len(all_cifs)):
    name, ext = os.path.splitext(all_cifs[file])
    name = name.split('/')
    #name = name[5].split('_')
    name = name[5].split('_')
    all_idx.append(name[1])

    
idx_lst = []
for i in idx:
    if i in all_idx:
        idx_lst.append(i)
        
# Creating a path_ID
string = 'Structure_'
string2 = '.cif'
name_idx = [string+str(idx_lst[i])+string2 for i in range(len(idx_lst))]

cifs = []
for idx in range(len(name_idx)):
    paths = os.path.join(f'database/{id_a}/{id_b}/CSP/sorted_structures/', name_idx[idx])
    cifs.append(paths)
```

# Extracting Intensity values

```python
target_data = pd.read_csv(f'database/{id_a}/{id_b}/CSP/All_high.csv', sep="; ", engine="python", header=None)
#target_data = target_data.iloc[1: , :]
target_data.head()
```

# Creating Final Dataframe and save it as CSV file

```python
nsample = 381
data = pd.DataFrame(
    np.concatenate([np.expand_dims(np.array(cifs), axis=1), target_data, np.expand_dims(np.array(idx_lst), axis=1)], axis=1), 
    columns=["path"] + [f"angle_{i*0.1+2.0}" for i in range(nsample)] + ["id"]
)
print(data.shape)
#data.to_csv(f"./dataset/{id_b}.csv", encoding="utf-8", index=False)
data.to_csv(f"./dataset/VUSDIX_zp2.csv", encoding="utf-8", index=False)
data.head()
```
